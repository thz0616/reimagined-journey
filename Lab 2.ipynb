```python
import pandas as pd
from scipy.stats import pearsonr

#load the dataset into pandas dataframe
df = pd.read_csv('data/customer_data.csv')

#clean and preprocess the data as necessary. for example, you might remove any missing values or outliners
df.dropna(inplace=True)
df = df[df [ 'age' ] > 18]

corr, p_value = pearsonr(df[ 'age' ], df[ 'purchase_frequency' ])
print("Correlation: ", corr)
print("P-value: ", p_value)
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['age'], df['purchase_frequency'])
plt.xlabel('Age')
plt.ylabel('Purchase Frequency')
plt.show()
```

```python
import numpy as np
#create a sample array of data
data = np.array([2,4,6,8,10])

#compute the mean of the data
mean = np.mean(data)

#compute the median of the data
median = np.median(data)

#compute the variance of data
variance = np.var(data)

#compute the standard deviation of data
stddev = np.std(data)

#compute the correlation between two arrays of data
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])
correlation = np.corrcoef(x,y)[0,1]

#perform a linear regression on two arrays of data
slope, intercept = np.polyfit(x, y, 1)
print("Mean: ", mean)
print("Median: ", median)
print("Variance: ", variance)
print("Standard Deviation: ", stddev)
print("Correlation: ", correlation)
print("Slope: ", slope)
print("Intercept: ", intercept)
```
